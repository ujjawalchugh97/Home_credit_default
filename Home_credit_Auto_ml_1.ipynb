### DataTypes of Columns

```python
x = df_train.columns.to_series().groupby(df_train.dtypes).groups
x
```

### Print out the columns to be passed as categorical

```python
for i in x:
    if(i=='int64'):
        for j in range(len(x[i])):
            if(len(df_train[x[i][j]].unique()) < 10):
                print ('\t\'' +x[i][j]+ '\''+ ": \'categorical\',")
    elif(i == 'object'):
        for j in range(len(x[i])):
            print ('\t\'' +x[i][j]+ '\''+ ": \'categorical\',")
    
```

```python
column_descriptions = {
    'SK_ID_CURR' : 'ignore', #We don't want it to consider ID as one of the inputs
    'TARGET': 'output', #Output Variable
    'FLAG_MOBIL': 'categorical',
	'FLAG_EMP_PHONE': 'categorical',
	'FLAG_WORK_PHONE': 'categorical',
	'FLAG_CONT_MOBILE': 'categorical',
	'FLAG_PHONE': 'categorical',
	'FLAG_EMAIL': 'categorical',
	'REGION_RATING_CLIENT': 'categorical',
	'REGION_RATING_CLIENT_W_CITY': 'categorical',
	'REG_REGION_NOT_LIVE_REGION': 'categorical',
	'REG_REGION_NOT_WORK_REGION': 'categorical',
	'LIVE_REGION_NOT_WORK_REGION': 'categorical',
	'REG_CITY_NOT_LIVE_CITY': 'categorical',
	'REG_CITY_NOT_WORK_CITY': 'categorical',
	'LIVE_CITY_NOT_WORK_CITY': 'categorical',
#   'FLAG_DOCUMENT_2': 'categorical', #Passing on this as categorical shows error for some reason
	'FLAG_DOCUMENT_3': 'categorical',
	'FLAG_DOCUMENT_4': 'categorical',
	'FLAG_DOCUMENT_5': 'categorical',
	'FLAG_DOCUMENT_6': 'categorical',
	'FLAG_DOCUMENT_7': 'categorical',
	'FLAG_DOCUMENT_8': 'categorical',
	'FLAG_DOCUMENT_9': 'categorical',
	'FLAG_DOCUMENT_10': 'categorical',
	'FLAG_DOCUMENT_11': 'categorical',
	'FLAG_DOCUMENT_12': 'categorical',
	'FLAG_DOCUMENT_13': 'categorical',
	'FLAG_DOCUMENT_14': 'categorical',
	'FLAG_DOCUMENT_15': 'categorical',
	'FLAG_DOCUMENT_16': 'categorical',
	'FLAG_DOCUMENT_17': 'categorical',
	'FLAG_DOCUMENT_18': 'categorical',
	'FLAG_DOCUMENT_19': 'categorical',
	'FLAG_DOCUMENT_20': 'categorical',
	'FLAG_DOCUMENT_21': 'categorical',
	'NAME_CONTRACT_TYPE': 'categorical',
	'CODE_GENDER': 'categorical',
	'FLAG_OWN_CAR': 'categorical',
	'FLAG_OWN_REALTY': 'categorical',
	'NAME_TYPE_SUITE': 'categorical',
	'NAME_INCOME_TYPE': 'categorical',
	'NAME_EDUCATION_TYPE': 'categorical',
	'NAME_FAMILY_STATUS': 'categorical',
	'NAME_HOUSING_TYPE': 'categorical',
	'OCCUPATION_TYPE': 'categorical',
	'WEEKDAY_APPR_PROCESS_START': 'categorical',
	'ORGANIZATION_TYPE': 'categorical',
	'FONDKAPREMONT_MODE': 'categorical',
	'HOUSETYPE_MODE': 'categorical',
	'WALLSMATERIAL_MODE': 'categorical',
	'EMERGENCYSTATE_MODE': 'categorical',
}

ml_predictor = Predictor(type_of_estimator='classifier', 
                         column_descriptions=column_descriptions)
```

### Train the predictor (It'll figure out the hyperparameters on its own and do a feature analytics as well)

```python
ml_predictor.train(df_train,
                   #model_names = ['ARDRegression', 'AdaBoostClassifier', 'AdaBoostRegressor', 'BayesianRidge', 'ElasticNet', 'ExtraTreesClassifier', 'ExtraTreesRegressor', 'GradientBoostingClassifier', 'GradientBoostingRegressor', 'Lasso', 'LassoLars', 'LinearRegression', 'LogisticRegression', 'MiniBatchKMeans', 'OrthogonalMatchingPursuit', 'PassiveAggressiveClassifier', 'PassiveAggressiveRegressor', 'Perceptron', 'RANSACRegressor', 'RandomForestClassifier', 'RandomForestRegressor', 'Ridge', 'RidgeClassifier', 'SGDClassifier', 'SGDRegressor', 'DeepLearningClassifier', 'DeepLearningRegressor', 'LGBMClassifier', 'LGBMRegressor'],
                   #Default model used : GradientBoostingClassifier
                   optimize_final_model='True'
)
```

```python
ml_predictor.predict(df_test)
```

```python
ans = ml_predictor.predict_proba(df_test)
ans
```

```python
df_sub = pd.DataFrame() # Create the final dataframe to submit
df_sub['SK_ID_CURR'] = df_test['SK_ID_CURR']
df_sub
```

```python
a = []
for i in ans:
    a.append(i[1])
a
```

```python
df_sub['TARGET'] = pd.DataFrame(a)
```

```python
df_sub = df_sub[:-1]
df_sub
```

```python
df_sub.set_index('SK_ID_CURR', inplace=True)
df_sub
```

```python
df_sub.to_csv('/home/ujjawal/Downloads/Home_Default_risk/final_sub.csv')
```
